```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 

df=pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)
print(df.head())

df_2020 = df[["Country Name", "2020"]].dropna()

top10 = df_2020.sort_values(by ="2020", ascending=False).head(10)
#BAR CHART
plt.figure(figsize=(12, 6))
sns.barplot(data=top10, x="Country Name", y="2020", palette="viridis")
plt.title("Top 10 Most Populated Countries in 2020")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#HISTOGRAM
plt.figure(figsize=(10, 6))
sns.histplot(df_2020["2020"], bins=30, kde=True, color="teal")
plt.title("Population Distribution Across Countries (2020)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```
